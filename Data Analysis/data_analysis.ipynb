```python
from sklearn.impute import SimpleImputer
```

```python
frequent_imputer = SimpleImputer(strategy='most_frequent')
macbook_data[['Design','Processing Speed','Graphic Performance','Expandability','Ease of Use','Portability', 'Battery', 'Screen','Cost Performance']] = frequent_imputer.fit_transform(macbook_data[['Design','Processing Speed','Graphic Performance','Expandability','Ease of Use','Portability', 'Battery', 'Screen','Cost Performance']])
macbook_data
```

```python
delete = ['Functionality', 'Ease of Use', 'Agility', 'Support']
window_data_delete = window_data.drop(columns=delete)
window_data_delete
```

```python
import matplotlib.pyplot as plt 

macbook_columns = ['Text','Satisfaction','Design','Processing Speed','Graphic Performance','Expandability','Ease of Use','Portability', 'Battery', 'Screen','Cost Performance']
macbook_labels = ['Text','Satisfaction','Design','Processing Speed','Graphic Performance','Expandability','Ease of Use','Portability', 'Battery', 'Screen','Cost Performance']

bins = 100
color = "blue"
grid = True

for i, col in enumerate(macbook_columns):
    plt.figure(figsize=(10, 4))
    macbook_data[col].hist(bins=bins, color=color, grid=grid)
    plt.title(f'Histogram of Macbook {macbook_labels[i]}')
    plt.xlabel(macbook_labels[i])
    plt.ylabel('Frequency')
    plt.show()
```
