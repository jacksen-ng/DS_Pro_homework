## Use The Graph To Do Data Analysis

## Graph Of Review's Macbook

```python
import matplotlib.pyplot as plt 

macbook_columns = ['Text','Satisfaction','Design','Processing Speed','Graphic Performance','Expandability','Ease of Use','Portability', 'Battery', 'Screen','Cost Performance']
macbook_labels = ['Text','Satisfaction','Design','Processing Speed','Graphic Performance','Expandability','Ease of Use','Portability', 'Battery', 'Screen','Cost Performance']

bins = 100
color = "blue"
grid = True

for i, col in enumerate(macbook_columns):
    plt.figure(figsize=(10, 4))
    macbook_data[col].hist(bins=bins, color=color, grid=grid)
    plt.title(f'Histogram of Macbook {macbook_labels[i]}')
    plt.xlabel(macbook_labels[i])
    plt.ylabel('Frequency')
    plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

data = macbook_data.drop('Text',axis=1)
correlation_matrix = data.corr()

plt.figure(figsize=(12, 8))

sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm')

plt.title('Correlation Matrix of MacBook Ratings')
plt.xlabel('Features')
plt.ylabel('Features')

plt.show()
```

## 考察

・スクレイピングデータ：スクレイピングによるデータ分析から、日本の消費者はMacBookに高い評価をしており、特に満足度が高いことが多い。特に、MacBookのグラフィック性能に関しては、最も高い評価を受けている。しかし、MacBookの拡張性については、Windowsパソコンと比較すると、それほど良いわけではない。また、Windowsパソコンのレビューを全体的に見ると、高評価は少ない。これは、Windowsの使い方が多様であり、各ブランドのWindowsパソコンには独自の拡張機能があるため、パソコン初心者にはあまり適していないと思われる

・ローカルデータ：今回の調査対象は、私の友人とデータサイエンスの先輩たちである。彼らは武蔵野大学のデータサイエンス学部の学生や卒業生で、この学部ではMacBookが推奨されている。その結果、多くの学生や教員がMacBookを利用し、その性能や使いやすさに高い評価をしていることがわかった。特に、その高速な処理能力と優れたグラフィック性能は、データ分析やプログラミング作業に適していると感じられている。一方で、Windowsパソコンについても調査を行いった。Windowsパソコンは、特にソフトウェアの互換性や柔軟性において高い評価を受けている。多くの学生は、特定のプログラミング言語やソフトウェアがWindows環境でしか利用できない場合、Windowsパソコンを選ぶ傾向にある。さらに、Windowsパソコンはカスタマイズの自由度が高く、ユーザーのニーズに合わせてハードウェアをアップグレードすることが可能である。

・全体的の考察：日本の消費者はMacBookに高い評価をしている。しかし、今回の分析ではWindowsパソコンに対して悪い印象を持つ可能性があるかもしれない。その原因としては、日本の消費者の多くがAppleの製品を主に使用しているためである。そのため、バイアスが発生する可能性があると考えている。また、データ量が少ないために、海外の消費者はWindowsパソコンに高い評価をしている可能性があると思思う。ブランド忠誠度の分析によると、日本の消費者はMacBookに高い忠誠度を示している。また、MacBookのグラフィック性能が高いため、デザイナーやビデオ編集者、音楽編集者、データサイエンティストなどの職業にはMacBookが推奨される。逆に、Windowsパソコンはソフトウェア開発者、ゲーム開発者、エンジニアなどに推奨されます。これはWindowsパソコンが様々な開発環境を提供しているためである。例えば、私の現在の研究ではアンリアルエンジンを使用し、テキストからモデルを生成するというテーマで研究している。アンリアルエンジンはMacBookでも起動可能だが、動作が重いと感じた。Windowsパソコンで起動すると、そのような重さは感じられない。
