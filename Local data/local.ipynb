## Translate the unique value

```python
translations_department = {
    'データサイエンス学科': 'Department of Data Science',
    'Bachelor of Psychology': 'Bachelor of Psychology',
    '航運管理': 'Shipping Management',
    'software': 'Software',
    '幼児教育学科': 'Department of Early Childhood Education',
    '写真学科': 'Department of Photography',
    'テキスタイル': 'Textile',
    '国際観光学部': 'Faculty of International Tourism',
    '社会': 'Social Sciences',
    'Library and Information Science': 'Library and Information Science',
    '総合情報学科': 'Department of Comprehensive Information Studies',
    '俳優': 'Actor',
    '言語学院': 'Language Institute',
    '日本語': 'Japanese Language',
    'Management of Culture Industries': 'Management of Culture Industries',
    '哲学科': 'Department of Philosophy',
    '法学': 'Law',
    'Psychology ': 'Psychology',
    '物理学科': 'Department of Physics'
}

df['Department'] = df['Department'].map(translations_department)

df.head()
```

```python
translations_macbook_type = {
    'Macbook Pro': 'Macbook Pro',
    'Macbook Air': 'Macbook Air'
}

translations_macbook_review = {
    'よかった Good': 'Good',
    'とてもよかった Very Good': 'Very Good',
    '普通 Normal': 'Normal',
    '最悪 Bad': 'Bad'
}

translations_windows_type = {
    'DELL': 'DELL',
    'ASUS': 'ASUS',
    'IDK': 'IDK',
    'Surface': 'Surface',
    'Lenovo': 'Lenovo',
    'Microsoft': 'Microsoft',
    'レッツノート': 'Let’s Note',
    'RAZER': 'RAZER',
    'lenovo': 'Lenovo',
    'HP': 'HP',
    'マウスコンピューター': 'Mouse Computer',
    '自作': 'Custom Built',
    'Huawei': 'Huawei',
    'ACER': 'ACER',
    'なし': 'None',
    'TUKUMO': 'TUKUMO',
    'LG': 'LG'
}

translations_windows_review = {
    'NO': 'NO',
    'よかった Good': 'Good',
    '普通 Normal': 'Normal',
    'とてもよかった Very Good': 'Very Good',
    '最初は困惑、慣れたら普通。': 'Confused at first, normal after getting used to it.',
    'なし': 'None',
    '最悪 Bad': 'Bad'
}

df['Type of Macbook'] = df['Type of Macbook'].map(translations_macbook_type)
df['Review of Macbook'] = df['Review of Macbook'].map(translations_macbook_review)
df["Type of Window's Laptop"] = df["Type of Window's Laptop"].map(translations_windows_type)
df["Review of Window's Laptop"] = df["Review of Window's Laptop"].map(translations_windows_review)

df.head()
```

## Create new csv file

```python
new_file_path = 'Review of Macbook and Window(local_data)_new.csv'  
df.to_csv(new_file_path, index=False)
```
